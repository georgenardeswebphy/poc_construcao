```python
import tensorflow as tf
import numpy as np
import cv2 as cv
```

```python
PONTO_NAMES = ["01", "10", "11"]

#for i in range(6):
for PONTO_NAME in PONTO_NAMES:
    vidcap = cv.VideoCapture('dataset/videos/{}_{}.mp4'.format(PONTO_NAME,3))
    success, image = vidcap.read()
    count = 0
    while success:
        # image = tf.image.resize(image, size=(224, 224)).numpy()     
        image = cv.resize(image, (224, 224))
        # image = cv.rotate(image, cv.ROTATE_180) 

        if count % 3 == 0:
            cv.imwrite("dataset/imagens/{}/{}{}_{}.png".format(PONTO_NAME,PONTO_NAME,count,3), image)     # save frame as JPEG file      

        success, image = vidcap.read()    

        print('Read a new frame: ', success)
        count += 1
```
